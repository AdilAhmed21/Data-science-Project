## Filling Null values with mean and mode

```python
df['lot_acres'] = df['lot_acres'].fillna(df['lot_acres'].mean())
```

```python
df['sqrt_ft'] = df['sqrt_ft'].fillna(df['sqrt_ft'].mean())
```

```python
df['kitchen_features'] = df['kitchen_features'].fillna(df['kitchen_features'].mode()[0])
```

```python
df['garage'] = df['garage'].fillna(df['garage'].mean())
```

```python
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mean())
```

```python
df['fireplaces'] = df['fireplaces'].fillna(df['fireplaces'].mean())
```

```python
df['floor_covering'] = df['floor_covering'].fillna(df['floor_covering'].mode()[0])
```

```python
df['HOA'] = df['HOA'].fillna(df['HOA'].mode()[0])
```

```python
plt.figure(figsize = (20,20))
pl = 1

for column in df:
    if pl<=14:
        ax = plt.subplot(4,4,pl)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize = 20)
    pl+=1
plt.show()
```
